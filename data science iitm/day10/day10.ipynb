```python
import random
import numpy as np
import pandas as pd
products=['chair','tablets','laptop','mobile phones','charger']
region=['east','west','north','south']

data={'order_id':range(1,1001,),
     'region_name':random.choices(region,k=1000),
     'product_name':random.choices(products,k=1000),
     'sales':np.random.randint(100,1000,1000),
     'quantity':np.random.randint(1,10,1000),
     'profit':np.random.randint(-1000,1000,1000),# 1 se 10 tak 1000 data
     'date_series': pd.date_range(start='2025-07-01', periods=1000)}

table=pd.DataFrame(data)
table
table.head(10)
```
