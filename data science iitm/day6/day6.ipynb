```python
#access modifiers public private and protected
class ATM_version_1:
    #protected _
    def __init__(self):
        print('object created successfully')
        self._pin=111#using the underscore beflre any variable we can make it private so it can not be accessed
        self._balance=2000

    def change_pin(self):
         self._pin = input('enter new pin')
         print('your pin has been changed ')
    def show_balance(self):
         print(f''' your balance current balance is rs {self._balance}''')
    def atm_performance_status(self):
         print('atm is working fine')
```

```python
s=ATM_version_1()
```

```python
s.show_balance()
```

```python
s._pin #we can still use the so called private variables
```

```python
#access modifiers public private and protected
class ATM_version_2:
    #private __
    def __init__(self):
        print('object created successfully')
        self.__pin=111 #using the underscore beflre any variable we can make it private so it can not be accessed
        self.__balance=2000

    def change_pin(self):
         self.__pin = input('enter new pin')
         print('your pin has been changed ')
    def show_balance(self):
         print(f''' your balance current balance is rs {self.__balance}''')
    def __atm_performance_status(self):#private method
         print('atm is working fine')
```
