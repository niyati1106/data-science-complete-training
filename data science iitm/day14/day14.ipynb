```python
df_titanic=sns.load_dataset('titanic')
df_titanic.sample()
```

```python
import matplotlib.pyplot as plt
sns.countplot(data=df_titanic,x='sex',hue='alive',palette=sns.color_palette('magma',2))#bivariate analysis
plt.title('titanic survival analysis bases on gender')
plt.show()
```

```python
#julias ai
```

```python
#sns.barplot
sns.barplot(data=df_titanic,x='class',y='fare',hue='alive',estimator='mean') #we can pass min max sum etc in estimator
#x is for ategorical and y is for numerical
plt.show
#the lines are( ci line) confidence interwal jitni badi bar hogi utna analysis challenging hoga
```

```python
# sns.barplot

sns.barplot(data = df_titanic, x= 'class', y = 'fare',hue = 'alive',errorbar=('ci',0))#by making error bar 0 we can remove that bar that for ci
# min, max, sum
plt.grid(color='g',alpha=.4)
plt.show()

# Confidence Interval:95%
```

```python
#sns.boxplot:to detect outliars and to check data distribution
#can check the skewness of the graph
#learn about boxplot for interview
```

```python
#sns.boxplot()
sns.boxplot(data=df_titanic,x='class',y='fare', hue='sex')
```

```python
#sns.boxplot()
plt.figure(figsize=(10,5))
sns.boxplot(data=df_titanic,y='class',x='fare', hue='sex',orient='h')
```

```python
#sns.violinplot
plt.figure(figsize=(10,5))
sns.violinplot(data=df_titanic,y='class',x='fare', hue='sex')
```

```python
df_tips=sns.load_dataset('tips')
df_tips.sample()
```

```python
sns.catplot(data=df_tips,x='time',y='total_bill',kind='bar',hue='sex',row='smoker',col='day') #kind se hum koisa bhi graph bana sakte hau without changing the functiojs
plt.show()
```

```python
sns.catplot(data=df_tips,x='time',y='total_bill',kind='violin',hue='sex',row='smoker',col='day') #kind se hum koisa bhi graph bana sakte hau without changing the functiojs
plt.show()
```

```python
sns.catplot(data=df_tips,x='time',y='total_bill',kind='box',hue='sex',row='smoker',col='day') #kind se hum koisa bhi graph bana sakte hau without changing the functiojs
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.lmplot(data=df_tips ,x='total_bill',y='tip',row='smoker',col='day',hue='sex')
plt.show()
```

```python
sns.relplot(data=df_tips,x='total_bill',y='tip',kind='scatter',hue='sex',row='smoker',col='day')
plt.show()
```

```python
df_iris=sns.load_dataset('iris')
df_iris.sample()
```

```python
sns.pairplot(data = df_iris)
plt.show()
```

```python
sns.pairplot(data = df_iris, hue = 'species')
plt.show()
```

```python
sns.pairplot(data = df_iris, hue = 'species',corner = True)
plt.show()
```

```python
# sns.kdeplot : Kernel density estimation

sns.kdeplot(data = df_iris, x = 'sepal_length', y = 'petal_width', hue = 'species')

# sns.scatterplot(data = df_iris, x = 'sepal_length', y = 'petal_width', hue = 'species')
plt.show()
plt.show()
```

```python
# sns.kdeplot : Kernel density estimation

sns.kdeplot(data = df_iris, x = 'sepal_length', y = 'petal_width', hue = 'species')
sns.scatterplot(data = df_iris, x = 'sepal_length', y = 'petal_width', hue = 'species')
plt.show()
plt.show()
```

```python
# sns.kdeplot : Kernel density estimation

sns.kdeplot(data = df_iris, x = 'sepal_length', y = 'sepal_width', hue = 'species')
sns.scatterplot(data = df_iris, x = 'sepal_length', y = 'sepal_width', hue = 'species')
plt.show()
```

```python
#joint plot

sns.jointplot(data = df_iris, x = 'sepal_length', y = 'sepal_width', hue = 'species',kind = 'hist')
plt.show()
```

```python
#joint plot

sns.jointplot(data = df_iris, x = 'sepal_length',y = 'sepal_width',kind = 'reg')
plt.show()
```
