```python
data={'study_hrs':range(1,11),
      'marks_obtained':[25,33,46,50,56,67,70,83,97,100]
     }

df=pd.DataFrame(data)
df

```

```python
# Import Module
from sklearn.linear_model import LinearRegression

# Object Model Creation
model = LinearRegression()


# x,y

x=df[['study_hrs']]
y=df['marks_obtained']

model.fit(x,y)
```

```python
x=[[5.5]]
m=model.coef_
c=model.intercept_

print(m,c)
```

```python
study_hrs=[[5.5]]
marks = m * study_hrs + c
print('predicted marks',marks[0][0])
```

```python
import warnings
warnings.filterwarnings('ignore')
study_hrs=input('enter study hours')
std_hr=[[float(study_hrs)]]
ans=model.predict(std_hr)
marks=round(ans[0],2)

print(f'''your predicted marks is:{marks}''')
```
